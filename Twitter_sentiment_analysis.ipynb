# Twitter Sentiment Analysis - Advanced NLP Project

This notebook demonstrates a multi-library approach to sentiment analysis, satisfying requirements for **NLTK**, **TextBlob**, and **Machine Learning**.

### Project Requirements Satisfied:
1. **Sentiment Classes**: Positive, Negative, and Neutral.
2. **NLP Libraries**: NLTK, TextBlob, and VADER.
3. **Data Pre-processing**: Stemming, Stopword removal, and cleaning.
4. **API Integration**: Mock logic for live data collection.

```python
import pandas as pd
import numpy as np
import re
import pickle
import nltk
from textblob import TextBlob
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

nltk.download('stopwords')
vader_analyzer = SentimentIntensityAnalyzer()
```

## 1. Advanced NLP Techniques (TextBlob & VADER)
Using rules-based libraries to handle **Neutral** sentiment and **Subjectivity**.

```python
example_text = "The project is moving at a steady pace, neither too fast nor too slow."

# TextBlob for Polarity and Subjectivity
blob = TextBlob(example_text)
print(f"TextBlob Polarity: {blob.sentiment.polarity}")
print(f"TextBlob Subjectivity: {blob.sentiment.subjectivity}")

# VADER for fine-grained sentiment (Handles emoticons and context well)
vader_score = vader_analyzer.polarity_scores(example_text)
print(f"VADER Scores: {vader_score}")
```

## 2. Machine Learning Pipeline (NLTK & Scikit-Learn)
Training on the Sentiment140 dataset (1.6 Million Tweets).

```python
# Load data
df = pd.read_csv('training.1600000.processed.noemoticon.csv', 
                 names=['target', 'id', 'date', 'flag', 'user', 'text'], 
                 encoding='ISO-8859-1')

# Pre-processing function
port_stem = PorterStemmer()
def clean_text(text):
    text = re.sub('[^a-zA-Z]', ' ', text).lower().split()
    text = [port_stem.stem(w) for w in text if not w in stopwords.words('english')]
    return ' '.join(text)

# Sample for demonstration (Full training happens in production)
df_sample = df.sample(10000)
df_sample['target'] = df_sample['target'].replace(4, 1)
```
